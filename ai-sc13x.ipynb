## 4. 다음 데이터의 covariance, correlation을 구하세요.

- x = [820, 760, 1250, 990, 1080, 1450, 1600]

- y = [0, 1, 7, 1, 0, 6, 4]

```python
x = np.array([820, 760, 1250, 990, 1080, 1450, 1600])
y = np.array([0, 1, 7, 1, 0, 6, 4])
```

```python
cov = np.cov(x,y)
cov
```

```python
cov[0,1]
```

```python
x= np.std(x, ddof=1)
y= np.std(y, ddof=1)
corr = cov[0,1]/(x*y)
corr
```

## 6. 위 데이터에 PCA 를 적용 후 scatterplot을 그리세요.

```python
df_1.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
af_pca = pca.fit(df_1)
af_pca
```

```python
print(pca.components_)
print(pca.explained_variance_)
```

```python
DF = pca.transform(df_1)
DF
```

```python
DF = pd.DataFrame(data=DF, columns = ['PC1', 'PC2'])
DF
```

```python
DF = pd.concat([DF, df['Country']], axis=1)
DF
```

```python
import seaborn as sns
sns.scatterplot(data=DF, x='PC1', y='PC2', hue='Country')
```
