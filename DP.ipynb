# 백트래킹

```python
def check(x):
    for i in range(x):
        if row[x] == row[i]:
            print('row[x] == row[i]')
            return False
            
        if abs(row[x] - row[i]) == x-i:
            print('abs(row[x] - row[i]) == x-i)')
            return False
    return True
```

```python
n = int(input())
```

```python
row = [0]*4
row
```

```python
result = 0
```

```python
def dfs(x):
    global result
    if x == n:
        print('dfs x == n')
        result +=1
    else:
        for i in range(n):
            row[x] = i
            print('row[x]=i',i)
            if check(x):
                dfs(x+1)
                print('---')
```
